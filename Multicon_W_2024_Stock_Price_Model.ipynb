```python
def convert_to_float(num):
    return float(num.replace(',', ''))
```

```python
x = pd.DataFrame(us_df['Nasdaq_100_Price'].map(convert_to_float))
y = us_df['Bitcoin_Price'].map(convert_to_float)

x_train = x[:1000]
y_train = y[:1000]

x_test = x[1000:]
y_test = y[1000:]

type(x_train)
# Total data = 1243
# 0 to 1000 --> training(1000)
# 1000 to 1243 --> testing(243)
```

```python
plt.scatter(x, y)
plt.xlabel('Nasdaq_100_Price')
plt.ylabel('Bitcoin_Price')
```

```python
import sklearn.linear_model as lm
```

```python
model1 = lm.LinearRegression()
model1.fit(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
print('Coefficient: ', model1.coef_)
print('Intercept: ', model1.intercept_)
```

```python
x_predict = pd.DataFrame([18004.70])
type(x_predict)
```

```python
model1.predict(x_predict)
```

```python
y_predicted = model1.predict(x_train)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_predicted, color = 'red')
plt.xlabel('Nasdaq_100_Price')
plt.ylabel('Bitcoin_Price')
```
